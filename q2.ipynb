```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
def f(x,y):
    return np.array(
        [-20*y[0]+10*y[1]+100, 10*y[0]-20*y[1]]
    )

def RK( a = 0.,
        b = 5.,
        y0=np.array([0.,0.]),
        m = 1000,
        n = 2
):
    h = (b - a) / m
    y = np.zeros((m + 1, n))
    y[0] = y0
    x = np.arange(a, b + h, h)
    tabela=np.zeros((m+1,n+1))
    tabela[0] =np.array([x[0], y[0][0], y[0][1]])
    for i in range(m):
        k1 = f(x[i], y[i])
        k2 = f(x[i] + h / 2, y[i] + k1 * h / 2)
        k3 = f(x[i] + h / 2, y[i] + k2 * h / 2)
        k4 = f(x[i] + h, y[i] + k3 * h)
        k = (k1 + 2 * (k2 + k3) + k4) / 6
        y[i + 1] = y[i] + h * k
        tabela[i+1]=np.array([x[i+1], y[i+1][0], y[i+1][1]])
    return tabela
```

```python
tabela1 = RK(m=1000)
tabela2 = RK(m=10)
tabela3 = RK(b=100,m=10**5)
```

```python
tabela1_data={
    'x_i':tabela1[:,0],
    'i1_i': tabela1[:,1],
    'i3_i': tabela1[:,2]
}
tabela1_frame= pd.DataFrame(tabela1_data)
tabela1_frame
```
